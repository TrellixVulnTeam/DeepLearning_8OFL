```python
# uniform distribution, you can't do np.array([2,2,2,2,2,2,2,2,2,2]) why? Cause you aren't
#distributed over a range where each value is the same frequency. You can do 2,3 or 2,3,4,5,6,7

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats  as stats

uniform_data = stats.uniform.rvs(size=100000,  # Generate 100000 numbers
                                 loc = 0,      # From 0 
                                 scale=10)     # To 10

pd.DataFrame(uniform_data).plot(kind="density",  # Plot the distribution
                               figsize=(9,9),
                               xlim=(-1,11))

plt.show()
print(uniform_data[0:20])
#because this plot is level from 0-10 each value has equal probability of occuring. 
```

<h6>Binomial distribution vs. Gaussian</h6>
Gaussian for continous, binomial for discrete. Binomial approximates gaussian if all events
are equally probable. If not then you can adjust. q) can you fit a gaussian to any binomial?
or can you fit a gaussian to a binomal? You can calculate the differences in coefficients as
a way to find difference between the 2 distributions. 



```python
#cdf:get probability value witin given range, can you use this to predict outliers? 

#pmf 


import numpy as np
import pandas as pd
import scipy.stats as stats

fair_coin_flips = stats.binom.rvs(n=10,        # Number of flips per trial
                                  p=0.5,       # Success probability
                                  size=10000)  # Number of trials

print( pd.crosstab(index="counts", columns= fair_coin_flips))

pd.DataFrame(fair_coin_flips).hist(range=(-0.5,10.5), bins=11)
plt.show()

biased_coin_flips = stats.binom.rvs(n=10,      # Number of flips per trial
                                  p=0.8,       # Success probability
                                  size=10000)  # Number of trials

# Print table of counts
print( pd.crosstab(index="counts", columns= biased_coin_flips))

# Plot histogram
pd.DataFrame(biased_coin_flips).hist(range=(-0.5,10.5), bins=11)
plt.show()

print();
result1 = stats.binom.cdf(k=5,        # Probability of k = 5 successes or less
                n=10,       # With 10 flips
                p=0.8)      # And success probability 0.8

print();
result2 = 1 - stats.binom.cdf(k=8,        # Probability of k = 9 successes or more
                    n=10,       # With 10 flips
                    p=0.8)      # And success probability 0.8
print("result1:",result1," result2:", result2)
```

<h6>Poisson Distribution</h6>
<p>Models time arrivals, can you use this to model time arrivals and use this to predict outliers?
Yes!!!!
</p>

$P(X=x) = \frac{\lambda^xexp(-\lambda)}{x!}$
<p>One way to use the PD is to model the time interval for a given number of events. This is
oppisite than what we are used to thinking of which is the number of events/time. 
(https://anomaly.io/anomaly-detection-poisson-distribution/) for an example. 
This is extremely useful because we do not have to wait for events to arrive, we can meaasure the
time interval and after the time interval we can determine the probability of a certain number
of events. This is much faster/efficient to implement in a visualization. 

</p>
