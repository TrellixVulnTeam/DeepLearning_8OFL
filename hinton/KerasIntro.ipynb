```python
import keras
import numpy as np

from keras.models import Sequential
from keras.layers.core import Dense
```

```python
#2 variable input
train_input = np.array([[0,0],[0,1],[1,0],[1,1]])
#output corresponding to input in train
output = np.array([[0],[0],[0],[1]])

train_input = train_input.astype("float32")
output = output.astype("float32")
#use this to implement a linear or logistic neuron? There is a functional API also
model = Sequential()

model.add(Dense(4,input_dim=2,activation="relu"))
model.add(Dense(1, activation="sigmoid"))
model.compile(loss='mean_squared_error', optimizer="Adam", metrics = ['binary_accuracy'])
model.fit(train_input, output,nb_epoch=550, verbose=2)
print (model.predict(train_input)) #these are the weights; check this manually? 


#weights:
#[[ 0.3878608 ]
# [ 0.42013025]
# [ 0.40104148]
# [ 0.52813023]]
#where are the second set of weights? 




```

```python
import keras
from keras.utils.np_utils import to_categorical

from keras.layers import Input, Dense
from keras.models import Model

input = Input(shape=[2])
probs = Dense(1, activation='sigmoid')(input)

model = Model(input=input, output=probs)
model.compile(optimizer='sgd', loss='binary_crossentropy')

weights = model.trainable_weights # weight tensors
#weights = [weight for weight in weights if model.get_layer(weight.name[:-2]).trainable] # filter down weights tensors to only ones which are trainable
gradients = model.optimizer.get_gradients(model.total_loss, weights) # gradient tensors

print (weights)

import keras as K

input_tensors = [model.inputs[0], # input data
                 model.sample_weights[0], # how much to weight each sample by
                 model.targets[0], # labels
                 0, # train or test mode
]

get_gradients = K.function(inputs=input_tensors, outputs=gradients)

nb_sample = 10
weights = model.trainable_weights # weight tensors

inputs = [np.random.randn(nb_sample, 2), # X
          np.ones(nb_sample), # sample weights
          np.random.randint(2, size=[nb_sample, 1]), # y
          0 # learning phase in TEST mode
]

print (zip(weights, get_gradients(inputs)))
```
